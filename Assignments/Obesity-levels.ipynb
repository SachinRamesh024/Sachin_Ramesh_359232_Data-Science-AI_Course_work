# Exploratory Data Analysis — Obesity Levels Dataset


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the local CSV file
fn = 'Obesity_DataSet.csv'
df = pd.read_csv(fn)
print('Loaded file:', fn)
print('Shape:', df.shape)
df.head(6)
```

```python
print('\n--- info() ---')
print(df.info())
print('\n--- describe() ---')
display(df.describe(include='all').T)
print('\n--- Missing values ---')
display(df.isnull().sum())
```

```python
# Identify likely target / obesity column
# Common column in this dataset: 'NObeyesdad' (obesity level label)
possible_targets = [c for c in df.columns if 'obes' in c.lower() or c.lower() in ('nobeyesdad','obesity','obesity_level','class')]
target_col = possible_targets[0] if possible_targets else df.columns[-1]
print('Using target column:', target_col)
plt.figure(figsize=(6,4))
if df[target_col].dtype == 'O' or df[target_col].nunique() < 12:
    sns.countplot(y=df[target_col], order=df[target_col].value_counts().index)
else:
    sns.histplot(df[target_col], bins=20)
plt.title('Target / Obesity-level distribution')
plt.show()
```

```python
# Univariate analysis for a selection of numeric columns
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# pick up to 8 useful numeric columns (height, weight, age, calories, etc. if present)
candidates = ['Age','Height','Weight','FCVC','NCP','CH2O','FAF','CALC','Calories','BMI','Daily_Calories']
cols_to_plot = [c for c in candidates if c in df.columns] + [c for c in num_cols if c not in candidates][:8]
seen = set(); cols = []
for c in cols_to_plot:
    if c not in seen:
        cols.append(c); seen.add(c)
for c in cols[:8]:
    plt.figure(figsize=(6,3))
    sns.histplot(df[c].dropna(), kde=False, bins=30)
    plt.title(f'Distribution: {c}'); plt.tight_layout(); plt.show()

# Boxplots for numeric columns to see outliers
for c in cols[:6]:
    plt.figure(figsize=(6,2.5))
    sns.boxplot(x=df[c].dropna())
    plt.title(f'Boxplot: {c}'); plt.show()
```

```python
# Categorical columns distribution (up to 6)
cat_cols = df.select_dtypes(include=['object','category']).columns.tolist()
for c in cat_cols[:6]:
    plt.figure(figsize=(6,3))
    sns.countplot(x=df[c], order=df[c].value_counts().index)
    plt.title(f'Count plot: {c}'); plt.xticks(rotation=45); plt.tight_layout(); plt.show()
```

```python
# Bivariate plots — relationships between two features and color by target
# Find two common numeric features and plot against each other
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(num_cols) >= 2:
    xcol, ycol = num_cols[0], num_cols[1]
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=df[xcol], y=df[ycol], hue=df[target_col], palette='tab10', alpha=0.7)
    plt.title(f'{xcol} vs {ycol} colored by {target_col}'); plt.show()

# Pairplot for small subset of numeric columns (limit to 5)
pair_cols = num_cols[:5]
if len(pair_cols) > 1:
    sns.pairplot(df[pair_cols + [target_col]] if target_col in df.columns and df[target_col].dtype == 'O' else df[pair_cols], corner=True, height=2.2)
    plt.suptitle('Pairplot (first numeric columns)', y=1.02)
    plt.show()
```

```python
# Correlation heatmap for numeric data
num_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(10,8))
corr = num_df.corr()
sns.heatmap(corr, cmap='coolwarm', center=0, annot=True, fmt='.2f')
plt.title('Correlation heatmap (numeric features)'); plt.show()
```

```python
# Group comparisons: mean values by obesity level (if target categorical)
if target_col in df.columns and df[target_col].dtype == 'O':
    grp = df.groupby(target_col).mean(numeric_only=True).T
    display(grp.head(10))
    plt.figure(figsize=(10,4))
    grp.iloc[:8].plot(kind='bar', subplots=False, figsize=(10,5))
    plt.title('Feature means by obesity level (first 8 features)'); plt.legend(bbox_to_anchor=(1.05,1)); plt.tight_layout(); plt.show()
```

```python
# If dataset has country or region columns, show counts (student-style)
possible_geo = [c for c in df.columns if c.lower() in ('country','region','city')]
if possible_geo:
    c = possible_geo[0]
    plt.figure(figsize=(8,3))
    sns.countplot(y=df[c], order=df[c].value_counts().index[:20])
    plt.title(f'Counts by {c} (top 20)'); plt.show()
```

## 🩺 Project Overview

This project focuses on analyzing the *Global Nutrition and Obesity Levels Dataset*, which contains data from various countries including Mexico, Peru, and Colombia. Using data visualization and basic statistical analysis, this study highlights the trends and correlations that contribute to rising obesity cases globally.
