```python
import os
import shutil
import random

# Paths
train_path = "dataset/train"
test_path = "dataset/test"
os.makedirs(test_path, exist_ok=True)

persons = ["sachin", "tom_cruise", "will_smith"]

for person in persons:
    train_person_path = os.path.join(train_path, person)
    test_person_path = os.path.join(test_path, person)
    os.makedirs(test_person_path, exist_ok=True)

    images = os.listdir(train_person_path)
    random.shuffle(images)

    # Take 20% for test
    test_size = int(len(images) * 0.2)
    test_images = images[:test_size]

    for img in test_images:
        shutil.move(os.path.join(train_person_path, img), os.path.join(test_person_path, img))

print("Train/Test split completed!")
```
