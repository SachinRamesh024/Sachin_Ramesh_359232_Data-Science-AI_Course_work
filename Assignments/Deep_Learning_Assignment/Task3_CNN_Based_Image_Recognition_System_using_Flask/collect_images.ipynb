```python
import cv2
import os
import time


person_name = "sachin"
save_path = f"dataset/train/{person_name}"
os.makedirs(save_path, exist_ok=True)

cap = cv2.VideoCapture(0)

img_count = 0
max_images = 200

print("Press 'q' to quit early")

while img_count < max_images:
    ret, frame = cap.read()
    if not ret:
        print("Camera not working")
        break

    # Fix mirror issue
    frame = cv2.flip(frame, 1)

    # Save image
    img_name = os.path.join(save_path, f"{img_count}.jpg")
    cv2.imwrite(img_name, frame)
    img_count += 1

    # Show count on screen
    cv2.putText(
        frame,
        f"Images: {img_count}/{max_images}",
        (20, 40),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2
    )

    cv2.imshow("Collecting Images", frame)

    time.sleep(0.4)  # controls capture speed

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

print("Image collection completed!")
```
